```python
datasetname = "cogensumm"
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import json
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the JSON file
file_path = f"../results/fizz_original_{datasetname}.json"

with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# ✅ "cut"이 "test"인 항목만 필터링
test_data = [entry for entry in data if entry.get("cut") == "test"]

# ✅ 점수 및 라벨 추출
scores = np.array([entry["score"] for entry in test_data])
labels = np.array([entry["label"] for entry in test_data])

# ✅ ROC-AUC 계산
roc_auc = roc_auc_score(labels, scores)

# ✅ ROC 곡선 좌표 계산
fpr, tpr, _ = roc_curve(labels, scores)

# ✅ ROC 곡선 그리기
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, marker='.', label=f'ROC-AUC = {roc_auc:.4f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (cut=test)')
plt.legend()
plt.grid()
plt.show()

# ✅ 최종 결과 출력
print(f"ROC-AUC Score (cut=test): {roc_auc:.4f}")
```

```python
from sklearn.metrics import precision_recall_curve, auc
import json
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the JSON file
file_path = f"../results/fizz_original_{datasetname}.json"

with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# ✅ "cut"이 "test"인 항목만 필터링
test_data = [entry for entry in data if entry.get("cut") == "test"]

# ✅ 점수 및 라벨 추출
scores = np.array([entry["score"] for entry in test_data])
labels = np.array([entry["label"] for entry in test_data])

# ✅ PR-AUC 계산 (모델)
precision, recall, _ = precision_recall_curve(labels, scores)
pr_auc = auc(recall, precision)

# ✅ 랜덤 분류기의 PR-AUC 계산 (baseline 비교용)
random_scores = np.random.rand(len(labels))
random_precision, random_recall, _ = precision_recall_curve(labels, random_scores)
random_pr_auc = auc(random_recall, random_precision)

# ✅ PR Curve Plot
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label=f'PR-AUC (Model) = {pr_auc:.4f}')
plt.plot(random_recall, random_precision, linestyle='--', color='red', label=f'PR-AUC (Random) = {random_pr_auc:.4f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve (cut=test)')
plt.legend()
plt.grid()
plt.show()

# ✅ 결과 출력
print(f"PR-AUC (Model, cut=test): {pr_auc:.4f}")
print(f"PR-AUC (Random baseline): {random_pr_auc:.4f}")
```
